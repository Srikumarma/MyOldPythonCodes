```python
# import Libraries 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load Data 
dataset=pd.read_csv('Employee_Salaries.csv')
print(dataset)
X=dataset.iloc[:,0:1].values
y=dataset.iloc[:,1].values
```

```python
#plot the datapoint 
plt.scatter(X,y,color='blue')
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Simple Linear Model 
from sklearn.linear_model import LinearRegression
linear_reg_model=LinearRegression()
linear_reg_model.fit(X,y)
```

```python
# visualize linear Regression line
plt.scatter(X,y, color='blue')
plt.plot(X,linear_reg_model.predict(X),color='green')
plt.title('Linear Regression')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Polynomial Linear Regression 
# converting independent features into polynomial features 
from sklearn.preprocessing import PolynomialFeatures
poly_ft=PolynomialFeatures(degree=4)
X_poly=poly_ft.fit_transform(X)
```

```python
# Regression model with polynomial features 
poly_reg_model=LinearRegression()
poly_reg_model.fit(X_poly,y)
```

```python
## visualize Polinomial linear Regression 
plt.scatter(X,y, color='blue')
plt.plot(X,poly_reg_model.predict(poly_ft.fit_transform(X)),color='green')
plt.title('Linear Regression')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Prediction comparison between Linear Regression Model and Polynomial Linear Regression Model
print(linear_reg_model.predict([[6.8]]))
print(poly_reg_model.predict(poly_ft.fit_transform([[6.8]])))
```
