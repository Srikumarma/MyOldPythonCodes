```r
#function to plot wssplot
wssplot <- function(data,nc=15,seed=1234)
{
    wss<-(nrow(data)-1)*sum(apply(data,2,var))
    for(i in 2:nc){
        set.seed(seed)
        wss[i] <- sum(kmeans(data, centers=i)$withinss)}
    plot(1:nc,wss,type="b",xlab = "no. of Clusters",ylab="within grp squares")
}
```

# Hierarchical clustering 

```r
# Using the dendrogram to find the optimal number of clusters
dendrogram = hclust(d = dist(head(new_Data1,50), method = 'euclidean'), method = 'ward.D')
plot(dendrogram,
     main = paste('Dendrogram'),
     xlab = 'Dust_Allergy',
     ylab = 'Smoking')
```

```r
# Using the dendrogram to find the optimal number of clusters
dendrogram = hclust(d = dist(new_Data1, method = 'euclidean'), method = 'ward.D')
plot(dendrogram,
     main = paste('Dendrogram'),
     xlab = 'Dust_Allergy',
     ylab = 'Smoking')
```

```r
# Using the dendrogram to find the optimal number of clusters
dendrogram = hclust(d = dist(new_Data2, method = 'euclidean'), method = 'ward.D')
plot(dendrogram,
     main = paste('Dendrogram'),
     xlab = 'Occp_Hazards',
     ylab = 'Short_Breath')
```
