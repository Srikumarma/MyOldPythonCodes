```r
#regular method
print(filter(select(flights, carrier, dep_delay), dep_delay>60))
#nesting
print(flights %>% 
select(carrier, dep_delay) %>%
filter(dep_delay>60))
```

```r
#regular calculation
x1=1:5
x2=2:6
sqrt((sum(x1-x2)^2))
#chaining
(x1-x2)^2 %>% sum() %>% sqrt()
```

```r
#select dep,arr,flight. filter dep > 700

print(flights %>% select(dep_time,arr_time,flight) %>% filter(dep_time > 700))
```

```r
#arrange method(to sort the data)
print(flights %>%
select(carrier, dep_delay) %>%
arrange(dep_delay))
```

```r
#arrange in descending order
print(flights %>%
select(carrier, dep_delay) %>%
arrange(desc(dep_delay)))
```

```r
#mutate(to add column)
flights %>%
select(distance, air_time) %>%
mutate(speed=distance/air_time*60)
```

```r
print(flights%>%select(distance,air_time,speed)%>%arrange(speed))
```
