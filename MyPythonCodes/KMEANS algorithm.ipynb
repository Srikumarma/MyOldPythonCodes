```r
x=c(185,170,168,179,182,188)
y=c(72,56,60,68,72,77)
df=data.frame(cbind(x,y))
plot(df)
```

```r
kmf=kmeans(df,2)
```

```r
kmf$size
kmf$centers
kmf$cluster
```

```r
cl=cbind(kmf$cluster)
library(cluster)
```

```r
clusplot(df,
        kmf$cluster,
        lines=0,
        shade=TRUE,
        color=TRUE,
        labels=2,
        plotchar=FALSE,
        span=TRUE,
          main = paste("Cluster"),
         xlab = 'x',
         ylab = 'y'
        )
```
