```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib Inline
from sklearn.cluster import KMeans
```

```python
iris_data = pd.read_csv("Iris.csv")
iris_data = iris_data.drop('Species', axis = 1)
iris_data = iris_data.drop('Id',axis = 1)
iris_data.head()
```

```python
# Split Dataset into train_test Purpose
X_train = iris_data[:130]
# X_train
```

```python
X_test = iris_data[131:]
# X_test
```

```python
#Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++' )
kmeans.fit(X_train)
```

```python
pred_kmeans = kmeans.predict(X_test)
```

```python
X_test['Predicted_Cluster_Number'] = pred_kmeans
X_test
```
