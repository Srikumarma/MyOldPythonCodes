```r
set.seed(1234)

# Generate data
x = rnorm(500)
#to fit  graphs in a row
par(mfrow = c(3, 2))

# Create a histogram
hist(x, freq = FALSE, main = "&quot;Histogram and density&quot;")

# Calculate density
dx = density(x)
# Add density
lines(dx, lwd = 2, col = "red")

# Plot the density without histogram
plot(dx, lwd = 5, col = "red",
main = "&quot;Density&quot;")
# Add the data-poins with noise in the X-axis
rug(jitter(x))

# Big bandwidth
plot(density(x, bw = 20), lwd = 2,
col = "red", main = "&quot;Too big bandwidth&quot;")

# Small bandwidth
plot(density(x, bw = 0.05), lwd = 2,
col = "red", main =" &quot;Too small bandwidth&quot;")

plot(dx, lwd = 2, main = "&quot;&quot;", xlab = "&quot;&quot;",
col = "red", xlim = c(-4, 6), ylim = c(0, 0.5))
polygon(dx, col = "red")
```
