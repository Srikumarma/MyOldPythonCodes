```r
slices=c(10,2,8,17,12)
lbls=c("US","UK","INDIA","JAPAN","ITALY")
pie(slices,labels=lbls,main="simple pie chart")
```

```r
#convert the numeric values into percentage
slices=c(10,2,8,17,12)
pct=round(slices/sum(slices)*100)
lbls=paste(c("US","UK","INDIA","JAPAN","ITALY")," ",pct,"%",sep="")
pie(slices,labels=lbls,main="simple pie chart",col = rainbow(5))
```

```r
#3d pie chart
slices=c(10,2,8,17,12)
pct=round(slices/sum(slices)*100)
lbls=paste(c("US","UK","INDIA","JAPAN","ITALY")," ",pct,"%",sep="")
pie3D(slices,labels=lbls,main="3D pie chart",col = rainbow(5),explode=0.0)
```
