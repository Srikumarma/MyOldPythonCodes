```r
counts=table(mtcars$gear)
barplot(counts,col = rgb(0.8,0.7,0.6,0.4))
counts
```

```r
#horizontal bar chart
barplot(counts,horiz = TRUE,col = rgb(0.2,0.1,0.7,0.6))
```

```r
#add title, axis names and legend
barplot(counts,col = c("red","blue","yellow"), legend=c("3rd Gear","4th Gear","5th Gear"),
        main="Gear bar plot",xlab = "improvement",ylab = "frequency",space = 5)
```

```r
#stack bar plot
counts=table(mtcars$vs,mtcars$gear)
barplot(counts,col = c("blue","yellow"), legend=rownames(counts),
        main="Car distribution by Gears and VS",xlab = "No. of Gears",ylab = "",space = 5)
```

```r
#grouped bar plot
counts=table(mtcars$vs,mtcars$gear)
barplot(counts,col = c("blue","yellow"), legend=c("vs","gears"),
        main="Car distribution by Gears and VS",xlab = "No. of Gears",ylab = "",beside=TRUE)
```
